```python
consumer_key=''
consumer_secret=''
access_token_key=''
access_token_secret=''
# consumer_key=''
# consumer_secret=''
# access_token_key=''
# access_token_secret=''
```

```python
def get_api():
    auth=tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token_key, access_token_secret)
    api = tweepy.API(auth, wait_on_rate_limit=True)
    return api
```

```python
api = get_api()
```
